```python
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import matplotlib.pyplot as plt
```

```python
vertices = {
    0: {'x': 1, 'y': 0, 'z': 0},
    1: {'x': 0, 'y': 1, 'z': 0},
    2: {'x': -1, 'y': 0, 'z': 0},
    3: {'x': 0, 'y': -1, 'z': 0},
    4: {'x': 0.5, 'y': 0, 'z': 1},
    5: {'x': 0, 'y': 0.5, 'z': 1},
    6: {'x': -0.5, 'y': 0, 'z': 1},
    7: {'x': 0, 'y': -0.5, 'z': 1},
}
```

```python
# 1. Describe this pyramid using the Indexed Face Set model.
faces = {
    'bottom': [0, 1, 2, 3],
    'top': [4, 5, 6, 7],
    'front': [0, 3, 7, 4],
    'back': [1, 2, 6, 5],
    'left': [2, 3, 7, 6],
    'right': [0, 4, 5, 1],
}

fig = plt.figure()
ax = Axes3D(fig)
ax.set_xlim3d(-2, 2)
ax.set_ylim3d(-2, 2)
ax.set_zlim3d(0, 2)

polygons = []
for face in faces.values():
    polygon = [[vertices[v]['x'], vertices[v]['y'], vertices[v]['z']] for v in face]
    polygons.append(polygon)
ax.add_collection3d(Poly3DCollection(polygons, edgecolors='black'))
plt.show()
```
